```python
import pandas as pd
import numpy as np

train_df = pd.read_csv('data_new.csv')
train_df.head()
```

```python
train_df.dtypes
```

```python
train_df.isnull().sum()
```

```python
train_df = train_df.dropna()
train_df.isnull().sum()
```

```python
for col in train_df.columns:
    display(train_df[col].value_counts())
```

```python
train_df['Dependents'].loc[train_df['Dependents']=='3+']=3
train_df['Dependents'].value_counts()
```

```python
train_df['Total_Income'] = train_df['ApplicantIncome'] + train_df['CoapplicantIncome']
train_df
```

```python
train_df.drop(columns=['ApplicantIncome','CoapplicantIncome','Loan_ID'], inplace=True)
```

```python
train_df
```

```python
train_df['Gender']= train_df['Gender'].map({'Male':0, 'Female':1})
train_df['Married']= train_df['Married'].map({'No':0, 'Yes':1})
train_df['Education']= train_df['Education'].map({'Graduate':0, 'Not Graduate':1})
train_df['Self_Employed']= train_df['Self_Employed'].map({'Yes':0, 'No':1})
train_df['Loan_Status']= train_df['Loan_Status'].map({'N':0, 'Y':1})
train_df['Property_Area']= train_df['Property_Area'].map({'Rural':0, 'Semiurban':1, 'Urban':2})

train_df.info()
```

```python
train_df['Dependents'] = pd.to_numeric(train_df['Dependents'])
```

```python
train_df.info()
```

```python
train_df.columns
```

```python
X = train_df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Total_Income']]
y = train_df.Loan_Status
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=4, random_state=5)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred_val = model.predict(X_val)
accuracy_score(y_val, pred_val)
```

```python
pred_train = model.predict(X_train)
accuracy_score(y_train, pred_train)
```

```python
import pickle
model_file = open('clf.pkl',mode='wb')
pickle.dump(model, model_file)
model_file.close()
```
